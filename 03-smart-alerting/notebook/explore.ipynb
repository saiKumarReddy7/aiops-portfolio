```python
import pandas as pd

data = pd.read_csv('../data/alerts.csv')
print("Shape: ",data.shape)
print("First 5 rows:")
print(data.head())
print("Alerts by severity:")
print(data['severity'].value_counts())
print(data['service'].value_counts())
```

```python
print("All Payment related alerts")
service = data[data['service']=="payment-service"].alert_message
for i,each in enumerate(service):
  print(f"Alert {i+1} : {each}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(data['alert_message'])
print("Alert messages converted to matrix:")
print(f"Shape: {X.shape}")
print(f"Each alert is now a vector of {X.shape[1]} numbers")
print(f"\nVocabulary size: {len(vectorizer.vocabulary_)} words")
print(f"\nSample vocabulary: {list(vectorizer.vocabulary_.keys())[:10]}")
```
